### Creating your first graph and running it in a session

```python
import tensorflow as tf
```

```python
x = tf.Variable(3, name="x")
y = tf.Variable(4, name="y")
f = x*x*y + y + 2
```

```python
sess = tf.Session()
sess.run(x.initializer)
sess.run(y.initializer)
result = sess.run(f)
print(result)
```

```python
sess.close()
```

```python
with tf.Session() as sess:
    x.initializer.run()
    y.initializer.run()
    result = f.eval() # session is automaticaly closed
```

```python
init = tf.global_variables_initializer() #prepare an init node

with tf.Session() as sess:
    init.run() # actually initialize all the variales
    restuly = f.eval()
```

```python
sess = tf.InteractiveSession()
init.run()
result = f.eval()
print(result)
```

```python
sess.close()
```

### Lifecycle of a node value

```python
w = tf.constant(3)
x = w + 2
y = x + 5
z = x * 3

with tf.Session() as sess:
    print(y.eval())
    print(z.eval())
```

If we want to evaluate y and z efficiently without evaluating w and x twice as in the prvious code yo must ask TensorFlow to evaluate both y and z in just one graph run, as shown in the following code:

```python
with tf.Session() as sess:
    y_val, z_val = sess.run([y,z])
    print(y_val)
    print(z_val)
```
